# Folder with noisy data

```python
import os
```

```python
from matplotlib import pyplot as plt
```

```python
txt_files = []
for root, dirs, files in os.walk("/content/kaggle_data/denoising-dataset-multiple-iso-levels/AlphaISP - Denoising Dataset/AlphaISP - Denoising Dataset"):
    for file in files:
        if file.endswith("AlphaISP_2592x1536_8bits_Scene10.png"):
            txt_files.append(os.path.join(root, file))
```

```python
for noisy_image in txt_files:
  print(noisy_image)
  plt.figure()
  plt.imshow(image_1)
plt.show()
```

# Noise Reduction

```python
import cv2

# Apply a Gaussian blur to the image
blurred_image = cv2.GaussianBlur(image_1, (5, 5), 0)

# Subtract the blurred image from the original image
noise_reduced_image = image_1 - blurred_image
```

```python
plt.figure()
plt.imshow(noise_reduced_image)
plt.title("Denoised Image")
```

# Change Detection

```python
image_2 = io.imread("/content/kaggle_data/denoising-dataset-multiple-iso-levels/AlphaISP - Denoising Dataset/AlphaISP - Denoising Dataset/PNG Data/2DNR Denoising/Noised/AlphaISP_2592x1536_8bits_Scene11.png")
diff_image = image_1 - image_2  # Calculate the difference between the two images
```

```python
plt.figure()
plt.imshow(diff_image)
plt.title("Change Detection")
```

# Masking

```python
mask = np.zeros_like(image_1)  # Create a mask with the same size as the image
mask[100:200, 100:200] = 1  # Set the pixels in the center of the mask to 1
masked_image = image_1 * mask  # Apply the mask to the image
```

```python
plt.figure()
plt.imshow(masked_image)
plt.title("Masking")
```
